#### Question 1: Some of the orders are stored in another csv file named bigkart_newsales. Read the csv file, store it in a data frame and add it to the bigkart_sales data frame. Find the total sales value of the category 'Office Supplies' after combining the dataframes

```python
import pandas as pd
```

```python
bigkart_sales=pd.read_csv('bigkart_sales.csv')
bigkart_newsales=pd.read_csv('bigkart_newsales.csv')
```

```python
bigkart_sales.head()
```

```python
bigkart_sales.shape
```

```python
bigkart_newsales.head()
```

```python
bigkart_newsales.shape
```

```python
data=[bigkart_sales, bigkart_newsales]
df=pd.concat(data)
df
```

```python
df.groupby('Category')['Sales'].sum()
```

#### Question 2: There are some duplicate rows in the data frame. Drop these rows and calculate the total sales value of the category Office Supplies.

```python
df1 = df[df.duplicated('Order ID')]
df1
```

There are seven duplicates in concatinated dataframe

```python
df.shape
```

```python
df2=df.drop_duplicates(keep=False)
```

```python
df2.shape
```

```python
df2.groupby('Category')['Sales'].sum()
```

#### Question 3: Find the most profitable category and sub category combination based on the net profit.

```python
# Finding the most profitable category and sub category combination based on the net profit in duplicates removed dataframe
```

```python
df3=df2.groupby(['Category','Sub-Category'])['Profit'].sum()
df3
```

In the dataframe which doesn't have duplicates, Technology > Phones is the most profitable based on the net profit

```python
# Finding the most profitable category and sub category combination based on the net profit in concatinated dataframe
```

```python
df4=df.groupby(['Category','Sub-Category'])['Profit'].sum()
df4
```

In the concatinated dataframe, Technology > Phones is the most profitable based on the net profit

#### Question 4: How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order order IDs in the data frame.

```python
df2['Test']=df2['Order ID'].apply(lambda x: x[8:])
df2.head(5)
```

```python
df2['Length']=df2['Test'].apply(lambda x: len(x))
df2.head(5)
```

```python
df2['Length'].value_counts()
```

In the dataframe which doesn't have duplicates, there are six order id's that are invalid

#### Question 5: Find the top 25 orders based on sales value and find the number of orders which belong to furniture category.

#### Question 6: Among the orders with sales>250 and profit>50, find the product name of the fourth highest order based on sales value.

```python
df7 = df2[(df2['Sales'] > 250) & (df2['Profit'] > 50)]
```

```python
df8=df7.sort_values(by=['Sales'], ascending=False)
df8
```

In the dataframe which doesn't have duplicates, 'Motorola Headset, with Caller ID' is the fourth highest order based on sales value.

#### Question 7: Remove the orders with negative profit by dropping the corresponding rows with negative Profit. Find the product that makes the lowest profit per Quantity in the Technology category.

```python
df9 = df2[df2['Profit']>0]
df10=df9[['Product Name','Category','Profit','Quantity']].groupby(['Category','Product Name'])[['Profit','Quantity']].min()
df10
```

In the dataframe which doesn't have duplicates, 'Logitech Keyboard, Programmable' makes the lowest profit per Quantity in the Technology category, making around 7.3 per Quantity
